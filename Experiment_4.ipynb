```python
from matplotlib import pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
la=['mca','mba','msw','Mcom']
ex=[0,0,0.3,0]
plt.pie(x,labels=la,explode=ex,startangle=135,shadow=True)
plt.show()
```

```python
import numpy as np
from PIL import Image
from scipy import ndimage
from matplotlib import pyplot as plt
a=Image.open("car.jpg")
b=np.array(a)
c=ndimage.zoom(b,(2,2,1))
d=Image.fromarray(c)
plt.imshow(d)
plt.show()
```
